```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Car_data.csv')

plt.scatter(df['Engine HP'], df['MSRP'])
plt.xlabel('Engine HP')
plt.ylabel('MSRP')
plt.title('Scatter Chart Graph')
plt.show()
```

```python
import pandas as pd
df = pd.read_csv('Car_data.csv')
x = df['Market Category']
y = df['Popularity']
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('Combo Plot')
ax.set_xlabel('Market Category')
ax.set_ylabel('Popularity')
ax.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Car_data.csv')

plt.scatter(df['city mpg'], df['Engine Cylinders'])
plt.xlabel('city mpg')
plt.ylabel('Engine Cylinders')
plt.title('Task 5A')
plt.show()
```
